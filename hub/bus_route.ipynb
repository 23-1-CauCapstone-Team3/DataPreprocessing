```python
prevRouteId = "null"
cnt = 0

for i in range(len(route)) :
    row = route.loc[i];
    
    if row["route_id"] != prevRouteId :
        cnt = 0;        
    
    route.loc[i]["order"] = cnt
    cnt += 1
    
    prevRouteId = row["route_id"];
```
